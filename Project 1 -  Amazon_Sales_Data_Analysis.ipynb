```python
# Convert the 'Units Sold' column to numeric, coercing errors to NaN
df['Units Sold'] = pd.to_numeric(df['Units Sold'], errors='coerce')

# Drop rows with NaN values in the 'Units Sold' column
df = df.dropna(subset=['Units Sold'])
```

```python
df.groupby('Item Type')['Units Sold'].mean().plot(kind='bar', color='blue')
plt.title('Avg. Units sold by Item Type')
plt.xlabel('Item Type')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.groupby('Order Date').sum()['Total Revenue'].plot(kind = 'line',color = 'blue')
plt.title('Total Revenue over time')
plt.xlabel('Order Date')
plt.ylabel('Total Revenue')
print('*'*20,'Total Revenue:',df['Total Revenue'].sum(),'*'*23)
plt.show()
```

# Month-wise Sales

```python
df['Order Month'] = df['Order Date'].dt.month
```

```python
# Convert 'Total Revenue' column to numeric type
df['Total Revenue'] = pd.to_numeric(df['Total Revenue'], errors='coerce')

# Group by 'Order Month' and sum 'Total Revenue', then plot the bar chart
df.groupby('Order Month')['Total Revenue'].sum().plot(kind='bar', color='blue')
plt.title('Total Revenue By Month')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()
```

# Year-wise Sales

```python
df['Order Year'] = df['Order Date'].dt.year
```

```python
# Convert 'Total Revenue' column to numeric type
df['Total Revenue'] = pd.to_numeric(df['Total Revenue'], errors='coerce')

# Group by 'Order Year' and sum 'Total Revenue', then plot the bar chart
df.groupby('Order Year')['Total Revenue'].sum().plot(kind='bar', color='blue')
plt.title('Total Revenue By Year')
plt.xlabel('Order Year')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# Convert 'Order Month' and 'Order Year' columns to categorical data types
df['Order Month'] = df['Order Month'].astype('category')
df['Order Year'] = df['Order Year'].astype('category')

# Group by 'Order Month' and 'Order Year', and sum the values
Yearly_Monthwise = df.groupby(['Order Month', 'Order Year'])[['Total Revenue', 'Total Cost', 'Total Profit']].sum().reset_index()
Yearly_Monthwise.head()
```

```python
sns.catplot(x ='Order Month', y = 'Total Profit', data = Yearly_Monthwise, kind = 'bar',color = 'blue', col = 'Order Year', col_wrap = 3)
```

# Heatmap is used to find correlation between features.

```python
# Select only the numeric columns for correlation calculation
numeric_columns = df.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation and plot the heatmap
sns.heatmap(numeric_columns.corr(method='pearson'), annot=True, cmap='YlOrBr')
```

Observations:
1.The above heatmap Unit Price and Unit Cost are Highly corelated.
2.The Unit Price moderately related to Total Revenue and Total Cost but less related to Total Profit.
3.Lastly their No relation between Order ID and rest other features.
